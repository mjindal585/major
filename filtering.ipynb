```python
import pandas as pd
```

```python
import numpy as np
```

```python
df=pd.read_csv('renamed.csv')
```

```python
df.head()
```

```python
df.dropna(axis=1, how='all')
df.dropna(axis=0, how='all')
```

```python
df.describe()
```

```python
df['GlycoHemoglobin'] = df['GlycoHemoglobin'].fillna(df['GlycoHemoglobin'].median())
df['SaggitalAbdominal'] = df['SaggitalAbdominal'].fillna(df['SaggitalAbdominal'].median())
df['ArmCircum'] = df['ArmCircum'].fillna(df['ArmCircum'].median())
df['GripStrength'] = df['GripStrength'].fillna(df['GripStrength'].median())
```

```python
df['Family_income'] = df['Family_income'].fillna(method='ffill')
df['Breast_fed'] = df['Breast_fed'].fillna(value = 1)

```

```python
df.loc[df['GlycoHemoglobin'] < 6.0, 'Diabetes'] = 0
df.loc[(df['GlycoHemoglobin'] >= 6.0) & (df['GlycoHemoglobin'] <= 6.4), 'Diabetes'] = 1
df.loc[df['GlycoHemoglobin'] >= 6.5, 'Diabetes'] = 2

df.head(10)
```

```python
df['Years_in_US'] = df['Years_in_US'].apply(lambda x: x if x > 0 else 0)
```

```python
df.head()
```

```python
df['Maritial_Status'] = df['Maritial_Status'].apply(lambda x: x if x > 0 else 0)
```

```python
df.head()
```

```python
df['Maritial_Status'] = df['Maritial_Status'].apply(lambda x: x if x < 1 else 1)
```

```python
df.head()
```

```python
df['Taking_Insulin'] = df['Taking_Insulin'].apply(lambda x: x if x > 0 else 0)
df['Taking_Insulin'] = df['Taking_Insulin'].apply(lambda x: x if x < 1 else 1)
df.head()
```

```python
df['Taking_Oral_Agents'] = df['Taking_Oral_Agents'].apply(lambda x: x if x > 0 else 0)
df['Taking_Oral_Agents'] = df['Taking_Oral_Agents'].apply(lambda x: x if x < 1 else 1)
df.head()
```

```python
df['Eyes_Affected'] = df['Eyes_Affected'].apply(lambda x: x if x > 0 else 0)
df['Eyes_Affected'] = df['Eyes_Affected'].apply(lambda x: x if x < 1 else 1)
df.head()
```

```python
df=df.drop('Martitial_Status',axis=1)
```

```python
df.head()
```

```python
df['Gender'] = df['Gender'].apply(lambda x: 'Male' if x == 1 else 'Female')
df.head()
```

```python
cols = ['Maritial_Status', 'Breast_fed','Taking_Insulin','Taking_Oral_Agents','Eyes_Affected','Diabetes']
df[cols] = df[cols].applymap(np.int64)
```

```python
df.head()
```

```python
df['Recent_BP'] = df['Recent_BP'].fillna(method='ffill')
df.head()
```

```python
df['Recent_BP'].value_counts()
```

```python
df['Recent_BP'] = df['Recent_BP'].apply(lambda x: x if x != 9999.0 else 120 )
```

```python
df.head()
```

```python
df['Recent_BP'].value_counts()
```

```python
df.to_csv('finalized.csv',index=False)
```
